```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('echirio.jpg',0)
high,wide = img.shape
plt.imshow(img,cmap = "gray")
plt.show()
img.shape
high2 = int(high/2)
wide2 = int(wide/2)

```

## nearest neighbor(最近傍補完)　回転

```python
theta = 1 / 4 * np.pi
img2 = np.zeros((high,wide))
near = np.zeros((high,wide))
                     
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        x = i - high2
        y = j - wide2
        di = int(x * np.cos(theta) - y * np.sin(theta)) + high2
        dj = int(x * np.sin(theta) + y * np.cos(theta)) + wide2
        if 0 <= di <  high and  0<= dj < wide:
            near[i,j] = img[di,dj]
        

plt.imshow(near,cmap = "gray")
plt.show()

print(near.shape)
```

## bilinear 回転

```python
theta = 3 / 4 * np.pi
img2 = img 
img2 = np.insert(img2,high,img2[high-1,],axis=0)
img2 = np.insert(img2,wide,img2[:,wide-1],axis=1)

bili = np.zeros((high+1,wide+1))

for i in range(img2.shape[0]):
    for j in range(img2.shape[1]):
        dx = i - high2
        dy = j - wide2
        di = dx * np.cos(theta) - dy * np.sin(theta) + high2
        dj = dx * np.sin(theta) + dy * np.cos(theta) + wide2
        dbi = int(di)
        dbj = int(dj)
        dis = np.zeros((2,2))
        w = np.zeros((2,2))
        d = 0
        if 0 <= dbi < high and 0 <= dbj < wide:
            for x in range(2):
                for y in range(2):
                    dis[x,y] = np.sqrt((i+x-di)**2 + (j+y-dj)**2)
                    d += dis[x,y]
            for x in range(2):
                for y in range(2):        
                    w[x,y] = dis[1-x,1-y] / d
                    bili[i,j] += w[x,y] * img2[dbi+x,dbj+y] 
                

plt.imshow(bili,cmap = "gray")
plt.show()

#cv2.imwrite("echiriomini.jpg",bili)

print(bili.shape)
```
